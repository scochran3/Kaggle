## Convert Sex and Embarked Features to Numbers

# Feature Engineering

## Goals
* Fill in the null values for age
* Create new features that could aid in prediction
* Look at if there is a way to get more information out of the name or ticket class (analyze the strings and look for patterns that could be used as features)

## Impute missing values for Age

```python
df['Age'].isnull().sum()
```

There are 263 (including test set as any transformation we apply will need to apply there as well) null values that we need to impute. One method would be to get rid of the row, but this is a lot of data to lose. Another option is to fill them with the overall mean. However, we can get more specific with our age estimate. Let's group the data on class and use the mean from those groups to impute the age. The logic here is that age is likely correlated with class as generally older people will have more money. So we find the average age per class and use that to fill in the missing age for each passenger.

```python
sns.boxplot(x='Pclass', y='Age', data=df)
```

```python
df_grouped = df.groupby(['Pclass'])[['Age']].mean()
ageFill = df_grouped.to_dict()

# Empty lists to hold ages
ages = []

# Loop over the data frame, if age is NOT null then add it to the ages list, if it IS NULL then look up the age approximation
# in the ageFill Dictionary
for i in range(len(df)):
    if pd.isna(df['Age'].iloc[i]):
        #  ageFill['Age'][df['Pclass'].iloc[i]] looks up in ageFill the average age based on the class from that row
        ages.append(int(ageFill['Age'][df['Pclass'].iloc[i]]))
    else:
        ages.append(df['Age'].iloc[i])
        
# Replace df['Age'] with our new list
df['Age'] = ages
```

## Name Feature
The name feature may seem useless at the start, but let's explore it's values to see if there is anything we use. For example, perhaps there are titles like 'Dr.' in the name that could be used as a feature. Perhaps there is nothing but it is worth checking.

```python
# Print all the names and browse through to see if there's anything we see we can use
for name in df['Name']:
    print (name)
```

### Things of Note
* It looks like we could make a "not married woman" category by looking at if their title has Ms. or Miss in it
* Some names have "Master" in it, which could indicate wealth and perhaps have a better survival chance. Not many names have this but we have a healthy aspect ratio (rows:columns) so we could add this feature without much penalty towards overfitting
* Major, Dr, and Sir are also in some names, perhaps we combine this with master as a "special title" feature or something similar


```python
# Create an UnmarriedWoman feature
df['UnmarriedWoman'] = df['Name'].apply(lambda x: 1 if "Miss" in x or 'Ms.' in x else 0)

# Create a special titles feature
specialTitles = ['Dr.', 'Master', 'Major', 'Sir']
df['SpecialTitle'] = df['Name'].apply(lambda x: 1 if any(n in x for n in specialTitles) else 0)

# Drop the name column now that we've made other features from it
df = df.drop(['Name'], axis=1)
```

## Family Size
We can check to see if the passenger has no family at all by checking if both SibSp and Parch are 0. The thinking here is perhaps families could help each other and thus passengers with families would have a higher survival rate.

```python
familySizes = []
for i in range(len(df)):
    if df['Parch'].iloc[i] == 0 and df['SibSp'].iloc[i] == 0:
        familySizes.append(1)
    else:
        familySizes.append(0)
df['SoloPassenger'] = familySizes
```

## Remove other useless features
Perhaps later we will re-evaluate but for now let's remove the ticket and cabin class. We do this because ticket seems hard to create some value out of and Cabin has mostly null values

```python
df = df.drop(['Ticket', 'Cabin'], axis=1)
```

```python
# Modifies Seaborn's countplot to add the bar heights label to make it easier to interpret
def countPlotWithLabels(feature):
    sns.countplot(df[feature], hue=df['Survived'], palette='husl')
    
    ax = plt.gca()

    for p in ax.patches:
        if not math.isnan(p.get_height()):
            ax.text(p.get_x() + p.get_width()/2., p.get_height(), '{}'.format(p.get_height()), 
                fontsize=12, color='black', ha='center', va='bottom')
        else:
            ax.text(p.get_x() + p.get_width()/2., 0, '{}'.format(p.get_height()), 
                fontsize=12, color='black', ha='center', va='bottom')

countPlotWithLabels('Sex')
```

# Building a model
Let's build a classifier with a bunch of different algorithms and then we will do parameter tuning on the one that performs the best with it's base settings

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, NuSVC, LinearSVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Separate out the data and the target
df_data = train[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'UnmarriedWoman', 'SpecialTitle', 'SoloPassenger']]
df_target = train['Survived']

# Create a dictionary that will hold our scores
scores = {}

# Build a LogisticRegression model
LR = LogisticRegression()
LR_scores = cross_val_score(estimator=LR, X=df_data, y=df_target, cv=5)
scores['Logistic Regression'] = [LR_scores.mean(), LR_scores.std()]

# Build a variety of SVC Models
SVC, NuSVC, LSVC = SVC(), NuSVC(), LinearSVC()
SVC_scores = cross_val_score(estimator=SVC, X=df_data, y=df_target, cv=5)
scores['SVC'] = [SVC_scores.mean(), SVC_scores.std()]

NuSVC_scores = cross_val_score(estimator=NuSVC, X=df_data, y=df_target, cv=5)
scores['NuSVC'] = [NuSVC_scores.mean(), NuSVC_scores.std()]

LSVC_scores = cross_val_score(estimator=LSVC, X=df_data, y=df_target, cv=5)
scores['SVC'] = [LSVC_scores.mean(), LSVC_scores.std()]

# Build a LDA and QDA model
LDA, QDA = LinearDiscriminantAnalysis(), QuadraticDiscriminantAnalysis()
LDA_scores = cross_val_score(estimator=LDA, X=df_data, y=df_target, cv=5)
QDA_scores = cross_val_score(estimator=QDA, X=df_data, y=df_target, cv=5)

scores['LDA'] = [LDA_scores.mean(), LDA_scores.std()]
scores['QDA'] = [QDA_scores.mean(), QDA_scores.std()]

# Ensemble tree models
RF, GB = RandomForestClassifier(), GradientBoostingClassifier()
RF_scores = cross_val_score(estimator=RF, X=df_data, y=df_target, cv=5)
GB_scores = cross_val_score(estimator=GB, X=df_data, y=df_target, cv=5)

scores['RF'] = [RF_scores.mean(), RF_scores.std()]
scores['GB'] = [GB_scores.mean(), GB_scores.std()]

scores
```

GradientBoost provided the best scores. Let's dive into how GB performed before going into parameter optimization. Let's look at the feature importances to see which features are most useful for prediction and a confusion matrix.

```python
GB.fit(df_data, df_target)
featureImportances = pd.DataFrame({'Feature':df_data.columns, 'Values':GB.feature_importances_}).set_index('Feature')
featureImportances.sort_values(by='Values', ascending=False).plot(kind='bar')
```

## Feature Importance
*Clearly most of the predictive power is in a handful of variables. We can play around with what features to include, but we can see that the first four or five features have most of the predictive power. Let's perform cross validation with a variety of models with a variety of features

```python
featureImportances = featureImportances.sort_values('Values', ascending=False)

# Create an empty list to hold the number of features and the mean accuracy score
scores, topNFeatures = [], []

'''
This will iterate over our features starting with the highest feature importance and add features one at a time. Each iteration
we keep track of the mean accuracy score. Once this is done we plot the scores to see which model had the best performance.
'''
for i in range(1, len(featureImportances)):
    features = df_data.columns[0:i]
    temp_data = df_data[features]
    GB2 = GradientBoostingClassifier()
    temp_scores = cross_val_score(estimator=GB2, X=temp_data, y=df_target)
    topNFeatures.append(i)
    scores.append(temp_scores.mean())

modelComplexity = pd.DataFrame({'Number Of Features':topNFeatures, 'Score':scores}).set_index('Number Of Features')
modelComplexity['Score'].plot(kind='bar', color='blue')
```

### Selecting our final model
* From the chart above we see that Gradient Boosting performed best with our first four features
* Our top 4 features were Fare, Age, Sex and PClass. Thus our final model will only have these features
* The last step will be to parameter tune our Gradient Boost model with a training set of these four predictors

## Parameter Optimization
For this we use scikit learns GridSearch which will look over all the possible combinations of parameters we specify. The three main parameters we will change are the learning rate, the number of trees and the max depth of the trees.

```python
from sklearn.model_selection import GridSearchCV

df_data_shrunk = df_data[['Fare', 'Age', 'Sex', 'Pclass']]

# Create a dictionary with what we want to search over
param_grid = {'n_estimators': [4000, 5000, 6000], 
              'learning_rate': [.01, .005, .02], 
              'max_depth': [1,2,3,4,5]}

# Train the GridSearch using all of our computing power!
grid = GridSearchCV(estimator=GB, param_grid=param_grid, n_jobs=-1)
grid_result = grid.fit(df_data_shrunk, df_target)
grid_result.best_score_
```

## Check what the best parameters from Grid Search were

```python
grid_result.best_params_
```

## Use these parameters in our final model

```python
GB = GradientBoostingClassifier(learning_rate = .01, max_depth=2, n_estimators=5000)
GB.fit(df_data_shrunk, df_target)
```

```python
predictions = GB.predict(test[['Fare', 'Age', 'Sex', 'Pclass']])
submission = pd.DataFrame({'PassengerId': test['PassengerId'].tolist(), 'Survived': predictions})
submission['Survived'] = submission['Survived'].astype(int)
submission = submission.set_index('PassengerId')

submission.to_csv('predictions.csv')
```
